```python
import joblib

pipeline = joblib.load('models/supervised_pipeline_simple.joblib')
```

```python
preprocessor = pipeline.named_steps['preprocess']
model = pipeline.named_steps['clf']
```

```python
predicted_stress = int(model.predict(X_processed)[0])
```
