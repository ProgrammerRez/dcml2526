```python
import pandas as pd

# Load dataset
df = pd.read_csv("system_metrics_binary.csv")

thresholds = {
    "cpu": df["cpu_ratio"].quantile(0.95),
    "ram": df["ram_ratio"].quantile(0.85),
    "disk": df["disk_ratio"].quantile(0.85)
}
# Your computed thresholds
THRESHOLDS = {
    "cpu": thresholds['cpu'],
    "ram": thresholds['ram'],
    "disk": thresholds['disk']
}


# Apply rule-based classification
df["pred_label"] = (
    (df["cpu_ratio"] > THRESHOLDS["cpu"]) |
    (df["ram_ratio"] > THRESHOLDS["ram"]) |
    (df["disk_ratio"] > THRESHOLDS["disk"])
).astype(int)

# Show counts
print("Label counts:")
print(df["pred_label"].value_counts())

print("\nLabel percentages:")
print(df["pred_label"].value_counts(normalize=True) * 100)
```
